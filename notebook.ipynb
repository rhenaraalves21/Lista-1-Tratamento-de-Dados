b) Identificar e listar as posi√ß√µes (c√©lulas) contendo valores faltantes.

```python
def listar_celulas_vazias(df, max_exemplos=10):
    """
    Lista as coordenadas (linha, coluna) de todas as c√©lulas vazias no DataFrame.

    Par√¢metros:
        df (pd.DataFrame): DataFrame a ser analisado.
        max_exemplos (int): N√∫mero m√°ximo de exemplos a exibir (None para todos).
    """
    celulas_vazias = []

    # Encontra as c√©lulas vazias
    for coluna in df.columns:
        linhas_vazias = df[df[coluna].isna()].index.tolist()
        for linha in linhas_vazias:
            celulas_vazias.append((linha, coluna))

    # Exibe o total e exemplos
    total_vazias = len(celulas_vazias)
    print(f"Total de c√©lulas vazias: {total_vazias}")

    if total_vazias > 0:
        print("\nExemplos (linha, coluna):")
        for i, (linha, coluna) in enumerate(celulas_vazias[:max_exemplos]):
            print(f"  ‚Üí Linha {linha + 1}, Coluna '{coluna}'")

    return celulas_vazias

vazias = listar_celulas_vazias(df, max_exemplos=5)
```

f) Identificar e listar as posi√ß√µes (c√©lulas) contendo valores que n√£o pertencem ao
dom√≠nio (tipo de dados) esperado.

```python
def validar_dominio(df, max_exemplos=10):
    # Express√£o regular para validar formatos de imagem/v√≠deo
    formato_regex = re.compile(r'^(image|video)\/[a-zA-Z0-9]+$')
    erros = []

    for idx, row in df.iterrows():
        message_type = row['message_type']
        has_url = row['has_media_url']
        has_media = row['has_media']
        media = row['media']
        media_type = row['media_type']
        media_url = row['media_url']

        # Valida√ß√£o gen√©rica para has_url
        if message_type == "Url":
            if not has_url:
                erros.append((idx, 'has_url', has_url, "Deve ser True quando message_type √© Url"))
        else:
            if has_url:
                erros.append((idx, 'has_url', has_url, "Deve ser False quando message_type n√£o √© Url"))

        # Valida√ß√£o para message_type = "Texto"
        if message_type == "Texto":
            if has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser False quando message_type √© Texto"))
            if not pd.isna(media_type):
                erros.append((idx, 'media_type', media_type, "Deve ser vazio quando message_type √© Texto"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type √© Texto"))
            if not pd.isna(media):
                erros.append((idx, 'media', media, "Deve ser vazio quando message_type √© Texto"))

        # Valida√ß√£o para message_type = "Imagem"
        elif message_type == "Imagem":
            if not has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser True quando message_type √© Imagem"))
            if not (isinstance(media_type, str) and formato_regex.match(media_type) and media_type.startswith('image/')):
                erros.append((idx, 'media_type', media_type, "Deve seguir o padr√£o image/formato"))
            elif not pd.isna(media):
                extensao = media_type.split('/')[1]
                if not str(media).lower().endswith(f'.{extensao.lower()}'):
                    erros.append((idx, 'media', media, f"Deve terminar com .{extensao} quando media_type √© {media_type}"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type √© Imagem"))

        # Valida√ß√£o para message_type = "Video"
        elif message_type == "Video":
            if not has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser True quando message_type √© Video"))
            if not (isinstance(media_type, str) and formato_regex.match(media_type) and media_type.startswith('video/')):
                erros.append((idx, 'media_type', media_type, "Deve seguir o padr√£o video/formato"))
            elif not pd.isna(media):
                extensao = media_type.split('/')[1]
                if not str(media).lower().endswith(f'.{extensao.lower()}'):
                    erros.append((idx, 'media', media, f"Deve terminar com .{extensao} quando media_type √© {media_type}"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type √© Video"))

        # Valida√ß√£o para message_type = "Url"
        elif message_type == "Url":
            if has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser False quando message_type √© Url"))
            if media_type != "url":
                erros.append((idx, 'media_type', media_type, "Deve ser 'url' quando message_type √© Url"))
            if pd.isna(media_url) or media_url == "":
                erros.append((idx, 'media_url', media_url, "N√£o pode ser vazio quando message_type √© Url"))
            if not pd.isna(media):
                erros.append((idx, 'media', media, "Deve ser vazio quando message_type √© Url"))
        else:
            erros.append((idx, 'message_type', message_type, "Tipo de mensagem desconhecido"))

     # Exibe o relat√≥rio de erros
    total_erros = len(erros)
    print(f"\nüîç RELAT√ìRIO DE VALIDA√á√ÉO")
    print(f"Total de erros encontrados: {total_erros}")

    if total_erros > 0:
        print("\nüö® Exemplos de erros encontrados:")
        for i, (linha, coluna, valor, msg) in enumerate(erros[:max_exemplos]):
            print(f"""
            ‚Üí Linha:    {linha}
            ‚Üí Coluna:   '{coluna}'
            ‚Üí Valor:    {valor}
            ‚Üí Problema: {msg}
            """)
    return erros

errors = validar_dominio(df, max_exemplos=5)
```

j) Crie uma coluna chamada ‚Äúsharings‚Äù contendo a quantidade de vezes que o texto
armazenado no atributo ‚Äútext‚Äù aparece no dataset.

```python
# Contar a frequ√™ncia de cada texto no dataset
contagem_textos = df['text_content_anonymous'].value_counts()

# Criar a coluna 'sharings' mapeando a contagem para cada texto
df['sharings'] = df['text_content_anonymous'].map(contagem_textos)

# Preencher NaN com 0 (caso haja textos vazios/missing)
df['sharings'] = df['sharings'].fillna(0).astype(int)

display(df[['text_content_anonymous', 'sharings']].sample(10))
```

k) Crie uma coluna chamada ‚Äúsentiment‚Äù contendo os valores: -1 para textos negativos,
0 para textos neutros e 1 para textos positivos.

```python
# Definir as condi√ß√µes e os valores correspondentes
conditions = [
    (df['score_sentiment'] >= 0.3) & (df['score_sentiment'] <= 1),   # Positivas (1)
    (df['score_sentiment'] > -0.3) & (df['score_sentiment'] < 0.3),  # Neutras (0)
    (df['score_sentiment'] >= -1) & (df['score_sentiment'] <= -0.3)  # Negativas (-1)
]

choices = [1, 0, -1]  # Valores correspondentes para cada condi√ß√£o

# Criar a coluna 'sentiment'
df['sentiment'] = np.select(conditions, choices, default=np.nan)
df['sentiment'] = df['sentiment'].fillna(0)

display(df[['score_sentiment', 'sentiment']].sample(10))
```

m) Identificar inconsist√™ncias entre os atributos (features).

```python
def identificar_inconsistencias(df):
    """
    Vers√£o robusta para identificar inconsist√™ncias entre atributos.
    Retorna um dicion√°rio com DataFrames contendo as linhas problem√°ticas.
    """
    inconsistencias = {}

    try:

        # Sentimento fora do dom√≠nio [-1, 1]
        if 'score_sentiment' in df.columns:
            mask = ~df['score_sentiment'].between(-1, 1, inclusive='both')
            inconsistencias['sentimento_invalido'] = df[mask].copy()

        # URLs inv√°lidas (apenas se a coluna existir e for string)
        if 'media_url' in df.columns and pd.api.types.is_string_dtype(df['media_url']):
            mask = df['media_url'].notna() & ~df['media_url'].str.startswith(('http://', 'https://'), na=True)
            inconsistencias['url_invalida'] = df[mask].copy()

        # Valores ausentes em ID
        if 'id_message' in df.columns:
            inconsistencias['chaves_ausentes'] = df[df['id_message'].isna()].copy()

    except Exception as e:
        print(f"Erro durante a verifica√ß√£o de inconsist√™ncias: {str(e)}")
        return None

    return inconsistencias

# Identificar inconsist√™ncias
inconsistencias = identificar_inconsistencias(df)

if inconsistencias is not None:
    for tipo, df_incon in inconsistencias.items():
        print(f"\n Inconsist√™ncia: {tipo.upper()}")
        print(f"   - Linhas afetadas: {len(df_incon)}")
        if not df_incon.empty:
            display(df_incon.sample(2))
else:
    print("N√£o foi poss√≠vel verificar inconsist√™ncias devido a erros.")
```
