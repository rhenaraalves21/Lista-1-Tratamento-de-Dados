b) Identificar e listar as posições (células) contendo valores faltantes.

```python
def listar_celulas_vazias(df, max_exemplos=10):
    """
    Lista as coordenadas (linha, coluna) de todas as células vazias no DataFrame.

    Parâmetros:
        df (pd.DataFrame): DataFrame a ser analisado.
        max_exemplos (int): Número máximo de exemplos a exibir (None para todos).
    """
    celulas_vazias = []

    # Encontra as células vazias
    for coluna in df.columns:
        linhas_vazias = df[df[coluna].isna()].index.tolist()
        for linha in linhas_vazias:
            celulas_vazias.append((linha, coluna))

    # Exibe o total e exemplos
    total_vazias = len(celulas_vazias)
    print(f"Total de células vazias: {total_vazias}")

    if total_vazias > 0:
        print("\nExemplos (linha, coluna):")
        for i, (linha, coluna) in enumerate(celulas_vazias[:max_exemplos]):
            print(f"  → Linha {linha + 1}, Coluna '{coluna}'")

    return celulas_vazias

vazias = listar_celulas_vazias(df, max_exemplos=5)
```

f) Identificar e listar as posições (células) contendo valores que não pertencem ao
domínio (tipo de dados) esperado.

```python
def validar_dominio(df, max_exemplos=10):
    # Expressão regular para validar formatos de imagem/vídeo
    formato_regex = re.compile(r'^(image|video)\/[a-zA-Z0-9]+$')
    erros = []

    for idx, row in df.iterrows():
        message_type = row['message_type']
        has_url = row['has_media_url']
        has_media = row['has_media']
        media = row['media']
        media_type = row['media_type']
        media_url = row['media_url']

        # Validação genérica para has_url
        if message_type == "Url":
            if not has_url:
                erros.append((idx, 'has_url', has_url, "Deve ser True quando message_type é Url"))
        else:
            if has_url:
                erros.append((idx, 'has_url', has_url, "Deve ser False quando message_type não é Url"))

        # Validação para message_type = "Texto"
        if message_type == "Texto":
            if has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser False quando message_type é Texto"))
            if not pd.isna(media_type):
                erros.append((idx, 'media_type', media_type, "Deve ser vazio quando message_type é Texto"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type é Texto"))
            if not pd.isna(media):
                erros.append((idx, 'media', media, "Deve ser vazio quando message_type é Texto"))

        # Validação para message_type = "Imagem"
        elif message_type == "Imagem":
            if not has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser True quando message_type é Imagem"))
            if not (isinstance(media_type, str) and formato_regex.match(media_type) and media_type.startswith('image/')):
                erros.append((idx, 'media_type', media_type, "Deve seguir o padrão image/formato"))
            elif not pd.isna(media):
                extensao = media_type.split('/')[1]
                if not str(media).lower().endswith(f'.{extensao.lower()}'):
                    erros.append((idx, 'media', media, f"Deve terminar com .{extensao} quando media_type é {media_type}"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type é Imagem"))

        # Validação para message_type = "Video"
        elif message_type == "Video":
            if not has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser True quando message_type é Video"))
            if not (isinstance(media_type, str) and formato_regex.match(media_type) and media_type.startswith('video/')):
                erros.append((idx, 'media_type', media_type, "Deve seguir o padrão video/formato"))
            elif not pd.isna(media):
                extensao = media_type.split('/')[1]
                if not str(media).lower().endswith(f'.{extensao.lower()}'):
                    erros.append((idx, 'media', media, f"Deve terminar com .{extensao} quando media_type é {media_type}"))
            if not pd.isna(media_url):
                erros.append((idx, 'media_url', media_url, "Deve ser vazio quando message_type é Video"))

        # Validação para message_type = "Url"
        elif message_type == "Url":
            if has_media:
                erros.append((idx, 'has_media', has_media, "Deve ser False quando message_type é Url"))
            if media_type != "url":
                erros.append((idx, 'media_type', media_type, "Deve ser 'url' quando message_type é Url"))
            if pd.isna(media_url) or media_url == "":
                erros.append((idx, 'media_url', media_url, "Não pode ser vazio quando message_type é Url"))
            if not pd.isna(media):
                erros.append((idx, 'media', media, "Deve ser vazio quando message_type é Url"))
        else:
            erros.append((idx, 'message_type', message_type, "Tipo de mensagem desconhecido"))

     # Exibe o relatório de erros
    total_erros = len(erros)
    print(f"\n🔍 RELATÓRIO DE VALIDAÇÃO")
    print(f"Total de erros encontrados: {total_erros}")

    if total_erros > 0:
        print("\n🚨 Exemplos de erros encontrados:")
        for i, (linha, coluna, valor, msg) in enumerate(erros[:max_exemplos]):
            print(f"""
            → Linha:    {linha}
            → Coluna:   '{coluna}'
            → Valor:    {valor}
            → Problema: {msg}
            """)
    return erros

errors = validar_dominio(df, max_exemplos=5)
```

j) Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto
armazenado no atributo “text” aparece no dataset.

```python
# Contar a frequência de cada texto no dataset
contagem_textos = df['text_content_anonymous'].value_counts()

# Criar a coluna 'sharings' mapeando a contagem para cada texto
df['sharings'] = df['text_content_anonymous'].map(contagem_textos)

# Preencher NaN com 0 (caso haja textos vazios/missing)
df['sharings'] = df['sharings'].fillna(0).astype(int)

display(df[['text_content_anonymous', 'sharings']].sample(10))
```

k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos,
0 para textos neutros e 1 para textos positivos.

```python
# Definir as condições e os valores correspondentes
conditions = [
    (df['score_sentiment'] >= 0.3) & (df['score_sentiment'] <= 1),   # Positivas (1)
    (df['score_sentiment'] > -0.3) & (df['score_sentiment'] < 0.3),  # Neutras (0)
    (df['score_sentiment'] >= -1) & (df['score_sentiment'] <= -0.3)  # Negativas (-1)
]

choices = [1, 0, -1]  # Valores correspondentes para cada condição

# Criar a coluna 'sentiment'
df['sentiment'] = np.select(conditions, choices, default=np.nan)
df['sentiment'] = df['sentiment'].fillna(0)

display(df[['score_sentiment', 'sentiment']].sample(10))
```

m) Identificar inconsistências entre os atributos (features).

```python
def identificar_inconsistencias(df):
    """
    Versão robusta para identificar inconsistências entre atributos.
    Retorna um dicionário com DataFrames contendo as linhas problemáticas.
    """
    inconsistencias = {}

    try:

        # Sentimento fora do domínio [-1, 1]
        if 'score_sentiment' in df.columns:
            mask = ~df['score_sentiment'].between(-1, 1, inclusive='both')
            inconsistencias['sentimento_invalido'] = df[mask].copy()

        # URLs inválidas (apenas se a coluna existir e for string)
        if 'media_url' in df.columns and pd.api.types.is_string_dtype(df['media_url']):
            mask = df['media_url'].notna() & ~df['media_url'].str.startswith(('http://', 'https://'), na=True)
            inconsistencias['url_invalida'] = df[mask].copy()

        # Valores ausentes em ID
        if 'id_message' in df.columns:
            inconsistencias['chaves_ausentes'] = df[df['id_message'].isna()].copy()

    except Exception as e:
        print(f"Erro durante a verificação de inconsistências: {str(e)}")
        return None

    return inconsistencias

# Identificar inconsistências
inconsistencias = identificar_inconsistencias(df)

if inconsistencias is not None:
    for tipo, df_incon in inconsistencias.items():
        print(f"\n Inconsistência: {tipo.upper()}")
        print(f"   - Linhas afetadas: {len(df_incon)}")
        if not df_incon.empty:
            display(df_incon.sample(2))
else:
    print("Não foi possível verificar inconsistências devido a erros.")
```
